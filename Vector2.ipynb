```python
Text='''Я помню чудное мгновенье:
Передо мной явилась ты,
Как мимолетное виденье,
Как гений чистой красоты.
В томленьях грусти безнадежной,
В тревогах шумной суеты,
Звучал мне долго голос нежный
И снились милые черты.
Шли годы. Бурь порыв мятежный
Рассеял прежние мечты,
И я забыл твой голос нежный,
Твои небесные черты.
В глуши, во мраке заточенья
Тянулись тихо дни мои
Без божества, без вдохновенья,
Без слез, без жизни, без любви.
Душе настало пробужденье:
И вот опять явилась ты,
Как мимолетное виденье,
Как гений чистой красоты.
И сердце бьется в упоенье,
И для него воскресли вновь
И божество, и вдохновенье,
И жизнь, и слезы, и любовь.'''
NlpText = nlp(Text)
print(NlpText)
```

```python
Stop = nlp.Defaults.stop_words
StopList = sorted(list(Stop))
```

```python
for StopWord in StopList:
    print(StopWord)
```

```python
FilteredWords = [token.text for token in NlpText if not token.is_stop]
FilteredText = ' '.join(FilteredWords)
print(FilteredText)
```

```python
NlpFilteredText = nlp(FilteredText)
VectorText = NlpText.vector
VectorFilteredText = NlpFilteredText.vector
```

```python
print(VectorText)
print(VectorFilteredText)
```

```python
from scipy.spatial.distance import cosine
```

```python
distance = cosine(VectorText, VectorFilteredText)
print(distance)
similar = NlpText.similarity(NlpFilteredText)
print(similar)
cosinDis = 1 - similar
print(cosinDis)
```

```python
Sentences = list(NlpText.sents)
for Sentence in Sentences:
    print(Sentence.text)
```

```python
SentenceVectors = {}
for i,Sentence in enumerate(Sentences):
    SentenceText = nlp(Sentence.text)
    FilteredSentenceText = ' '.join([token.text for token in SentenceText if not token.is_stop])
    print(FilteredSentenceText)
    FilteredSentenceVector = nlp(FilteredSentenceText).vector
    SentenceVectors[f"sentence_{i+1}"] = FilteredSentenceVector 
    print(FilteredSentenceVector)
```
